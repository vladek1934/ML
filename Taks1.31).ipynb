```python
import re
import numpy
import math
import matplotlib
import matplotlib.pyplot
from scipy.optimize import minimize

def f(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)


def f2(x):
    results = numpy.zeros(shape=(len(x)))
    for i in range(len(x)):
        results[i] = f(x[i])
    return results

def solve(x):
    n = len(x);
    points = []
    coefficients = [[0 for j in range(n)] for i in range(n)];
    
    for i in range(n): 
        points.append(f(x[i]))
        for j in range(n):        
            coefficients[i][j] =  x[i] ** j # setting up the coefficient matrix
        
    return(numpy.linalg.solve(coefficients, points))

a = [1, 15]
b = [1, 8, 15]
c = [1, 4, 10, 15]

sol1=solve(a)
sol2=solve(b)
sol3=solve(c)
print(sol1)
print(sol2)
print(sol3)

def plotpoints(sol, length):
    y = numpy.zeros(shape=(len(length)))
    for i in range(len(length)):
        for j in range(len(sol)):
                y[i] += sol[j] * (length[i] ** j)
    return y;

length = numpy.arange(1, 30, .2)
                    
y = list(map(f, length))
y2 = plotpoints(sol1, length)
y3 = plotpoints(sol2, length)
y4 = plotpoints(sol3, length)
matplotlib.pyplot.plot(length, y)
matplotlib.pyplot.plot(length, y2)
matplotlib.pyplot.plot(length, y3)
matplotlib.pyplot.plot(length, y4)

x = [0, 1, 2, 5, 8, 11, 13, 15, 20, 23, 25,25.88,25.9, 26, 27, 30]
print("\n")
print(f2(x))
print("\n")
for i in range(len(x)):
    res = minimize(f, x[i], method='BFGS')
    print(res.x,f(res.x))
```
