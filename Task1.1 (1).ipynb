```python
import re;
import math;
import scipy.spatial;
file = open("C:\\Users\\Angi\\Desktop\\TasksML\\sentences.txt")
content = []
for line in file:   
    line = re.split('[^a-z]', line.lower()) 
    line = list(filter(None, line)) # Remove all empty strings    
    content.append(line)   
#print (content); #check
TLines = len(content); #remember that for future use
print("Total lines count = " , TLines)
print('\n');

file.seek(0)

words = re.split('[^a-z]', file.read().lower()) #convert list to string
filtered = list(filter(None, words)) #remove whitespaces using a filter
print('\n');

#print (filtered);  #check
print("Total words count = " , len(filtered))
print (filtered.count("cat"));#check
print (filtered.count("")); #check
unique_words = set(words) #create a set of unique words
unique_words.remove('')#for some reason there is one '' that was not deleted, we delete it\
print('\n');

dictionary = list(unique_words)
print (dictionary) #check
UWords = len(unique_words) #for future use
print("Unique words count = " , UWords)
print (filtered.count("cat"));
print (filtered.count(""));
print('\n');    
    
a = [[0 for j in range(UWords)] for i in range(TLines)];
for i in range(TLines):
    for j in range(UWords):
        a[i][j] = content[i].count(dictionary[j])
           
for row in a:
    print(' '.join([str(elem) for elem in row]));
    print("\n")
    
def cosine_dist(x, y):
    Dot = 0
    A = 0
    B = 0
    for i in range(len(x)): 
        Dot += x[i] * y[i]
        A += x[i] * x[i] 
        B += y[i] * y[i]
    return 1 - (Dot / (math.sqrt(A) * math.sqrt(B)))

dists = []
min = 1;
min2 = 1;
for row in a:
    #dists.append(scipy.spatial.distance.cosine(a[0], row))
    dists.append(cosine_dist(a[0], row))
    
for n in dists:
    if min>n and n!=0:
        min = n;

for n in dists:
    if min2>n and n!=min and n!=0:
        min2 = n;
line = dists.index(min)+1;
line2 = dists.index(min2)+1;
print(dists)
print(min, min2)
print("The nearest sentences are" , line, line2)
```
