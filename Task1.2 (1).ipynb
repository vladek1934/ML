```python
import re
import numpy
import math
import matplotlib
import matplotlib.pyplot
def f(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)\

def solve(x):
    n = len(x);
    points = []
    coefficients = [[0 for j in range(n)] for i in range(n)];
    
    for i in range(n): 
        
        points.append(f(x[i]))
        
        for j in range(n):        
            coefficients[i][j] =  x[i] ** j # setting up the coefficient matrix
        
    return(numpy.linalg.solve(coefficients, points))

a = [1, 15]
b = [1, 8, 15]
c = [1, 4, 10, 15]

sol1=solve(a)
sol2=solve(b)
sol3=solve(c)
print(sol1)
print(sol2)
print(sol3)

def plotpoints(sol, length):
    y = numpy.zeros(shape=(len(length)))
    for i in range(len(length)):
        for j in range(len(sol)):
                y[i] += sol[j] * (length[i] ** j)
    return y;

length = numpy.arange(1, 15, .05)
                    
y = list(map(f, length))
y2 = plotpoints(sol1, length)
y3 = plotpoints(sol2, length)
y4 = plotpoints(sol3, length)
matplotlib.pyplot.plot(length, y)
matplotlib.pyplot.plot(length, y2)
matplotlib.pyplot.plot(length, y3)
matplotlib.pyplot.plot(length, y4)
```
